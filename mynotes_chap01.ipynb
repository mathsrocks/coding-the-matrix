## 1.3 Abstracting over _fields_

In programming languages, use of the same name (e.g. `+`) for different procedures operating on values of different data types is called _overloading_.

```python
def solve1(a, b, c): return (c - b)/a
```

```python
solve1(10, 5, 30)
```

The remarkable thing, however, is that the same procedure can be used to solve equations involving complex numbers.

```python
solve1(10 + 5j, 5, 20)
```

The procedure works even with complex arguments because the correctness of the procedure does not depend on what kind of numbers are supplied to it; it depends only on the fact that the _divide_ operator is the inverse of the _multiply_ operator and the _subtract_ operator is the inverse of the _add_ operator.

The power of this idea goes well beyond this simple procedure. Much of linear algebra concepts, theorems and procedures work not just for real numbers but also for complex numbers and for other kinds of numbers as well. The strategy for achieving this is simple:

* The concepts, theorems and procedures are stated in terms of the arithmetic operators `+`, `-`, `*` and `/`.
* They assume only that these operators satisfy certain basic laws, such as commutativity and distributivity.

Because the concepts, theorems and procedures rely only on these basic laws, we can "plug in" any system of numbers, called a _field_. 1 Different fields arise in different applications.

The author illustrates the generality of linear algebra with three different fields:
* $\mathbb R$
* $\mathbb C$
* $GF(2)$, a field that consists of 0 and 1.

In object-oriented programming, one can use the name of a class to refer to the set of instances of that class. In mathematics, one uses the name of the field, e.g. $\mathbb R$ or $GF(2)$, to refer to the set of values.

### 1.4.1 The absolute value of a complex number

The _absolute value_ of a complex number $z$, written as $\left|z\right|$ (and, in Python, `abs(z)`) is the distance from the origin to the corresponding point in the complex plane.

By the Pythagorean Theorem, $\left|z\right| = (z.real)^2 + (z.imag)^2$

```python
abs(3 + 4j)
```

```python
abs(1 + 1j)
```

**Definition 1.4.2**: The _conjugate_ of a complex number $z$, written $\bar z$, is defined as $z.real - z.imag$. In Python, we write `z.conjugate()`.

### 1.4.2 : Adding complex numbers

The transformation defined by complex number addition is called a _translation_. A translation has the form

$$

where $z_0$ is a complex number. Translations can take the picture anywhere in the complex plane.

#### Complex numbers as arrows

It is helpful to visualise a translation $f(z)$ by an arrow. The tail of the arrow is located at any point $z$ in the complex plane; the head of the arrow is then located at the point $f(z)$, the translation of $z$. Of course, this representation is not unique.

Since a translation has the form $f(z) = z_0 + z$, we represent the translation by the complex number $z_0$. It is therefore appropriate to represent the complex number $z_0$ by an arrow. Again, the representation is not unique. 

#### Composing translations, adding arrows

Let $f(z) = z_1 + z$ and $f(z) = z_2 + z$ be two translations. Then their composition is also a translation:

\begin{align}
    (f_2\circ f_1)(z) & = f_2(f_1(z))\\
                      & = f_2(z_1 + z)\\
                      & = z_2 + z_1 + z
\end{align}

and is defined by $z\mapsto (z_2 + z_1) + z$. The idea that two translations can be collapsed into one is better illustrated by arrows.

### 1.4.7 Euler's formula

Leonhard Euler is one of the most prolific mathematicians throughout the history. Euler's formula states that, for any **real** number $\theta$, $e^{\bs i \cdot \theta}$ is the point $z$ on the unit circle with _argument_ $\theta$ where $e$ is the famous transcendental number 2.72.

**Example 1.4.16**: The point $-1 + 0\bs i$ has argument $\pi$. Plugging $\pi$ into Euler's formula yields the surprising equation $e^{\bs i\pi} + 1 = 0$ or equivalently $e^{\pi\sqrt{-1}} = -1$.

### 1.4.8 Polar representation for complex numbers

$$
z = r e^{\theta\bs i}
$$

$r$ and $\theta$ are the polar coordinates of $z$. In the context of complex numbers, we define the _argument_ of $z$ to be $\theta$ and the _absolute value_ of $z$ (written $\left|z\right|$) to be $r$.

### 1.4.9 The FIrst Law of Exponentiation

When powers multiply, their exponents add:

$$
e^u e^v = e^{u + v}
$$

This rule helps us understand how to _rotate_ a complex number $z$. First we write the number in its polar form:

$$
z = r e^{\theta\bs i}
$$

where $r = \left|z\right|$ and $\theta = arg(z)$.

### 1.4.10 Rotation by $\tau$ radians

Let $\tau$ be a number of radians. The rotation of $z$ by $\tau$ should have the same absolute value as $z$ but its argument should be $\tau$ more than that of $z$, i.e. it should be $r e^{(\theta + \tau)\bs i}$:

$$
\begin{align}
r e^{(\theta + \tau)\bs i} & = r e^{\theta\bs i} e^{\tau \bs i}\\
                           & = z e^{\tau \bs i}.
\end{align}
$$

Thus the function that rotates by $\tau$ is simply
$$
f(z) = z e^{\tau\bs i}.
$$

## 1.5 Playing with _GF_(2)

_GF_(2) is short for _Galois Field 2_.

Addition is modulo 2. It is equivalent to **exclusive-or**. In particular, $1 + 1 = 0$.

Subtraction is identical to addition. The negative of 1 is again 1 and the negative of 0 is again 0.

Multiplication in _GF_(2) is just like ordinary multiplication of 0 and 1: multiplication by 0 yields 0, and 1 times 1 is 1. We can divide by 1 but dividing by zero is illegal (as usual).

The author provides a module, _`GF2`_, with a very simple implementation of _GF_(2). It defines a value, `one`, that acts as the element 1 of _GF_(2). Ordinary zero plays the role of the element 0 of _GF_(2).  (For visual consistency, the module defines `zero` to be the value 0.)

### 1.5.1 Perfect secrecy revisited

The encryption method is just _GF_(2) addition in disguise!

For each plaintext $p \in GF(2)$, the function $k \mapsto k + p$ (mapping _GF(2)_ to _GF(2)_) is invertible (hence one-to-one and onto). Therefore, when the key $k$ is chosen uniformly at random, the cyphertext is also distributed _uniformly_. This shows that the scheme achieves perfect secrecy.g 

**Using integers instead of _GF_(2)**

Why couldn't we use, say, ordinary integers instead of _GF_(2)? After all, for each $x\in \mathbb Z$, the function $y \mapsto x + y$ mapping $\mathbb Z$ to $\mathbb Z$ is also invertible. The eason this does not work as a cryptosystem is that there is **no** _uniform distribution_ over $\mathbb Z$, the first step - choosing a key - is impossible.

We can infer that the system has perfect secrecy.

The value your procedure initially assigns to `current` turns out to be the return value in the case when the input list `L` is empty. This provides us insight into how the answer should be defined in that case. Note: You are not allowed to use Python built-in procedures `sum(·)` and `min(·)`.

**Problem 1.7.6:**

```python
import math

def myMin(L):
    current = math.inf
    for x in L:
        if x < current: current = x
    return current
```

**Problem 1.7.7:**

```python
def myConcat(L):
    current = ""
    for x in L:
        current += x
    return current
```

**Problem 1.7.8:**

```python
def myUnion(L):
    current = set()
    for x in L:
        current = current | x
    return current
```

In each of the above problems, the value of `current` is combined with an element of `myList` using some operation $\diamond$. For the procedure to return the correct result, `current` should be initialised with the _identity element_ for the operation $\diamond$, i.e. the value $i$ such that $i\diamond x = x$ for any value $x$.

It is a consequence of the procedure structure that, when the input list is empty, the output value is the initial value of `current` (since in this case the body of the loop is never executed). It is convenient to define this to be the correct output!

Answer: The fact that translation comes before rotation and scaling means 
$$
2e^{-\frac{\pi}{2}\bs i}(z + (1 + 1\bs i)) = 2e^{-\frac{\pi}{2}\bs i}z + 2e^{-\frac{\pi}{2}\bs i}(1 + 1\bs i)
$$
Hence, $a = 2e^{-\frac{\pi}{2}\bs i}$ and $b = 2(1 + 1\bs i)e^{-\frac{\pi}{2}\bs i}$.

b. Scale the real part by two and the imaginary part by three, then rotate by forty-five degrees counterclockwise, and then translate down two units and left three units.

Answer: In my view, there is no way to achieve this transformation under the current infrastructure. We are unable to scale the real and imaginary parts of a complex number by different scales. That is when multiplying a complex number with the following

* A (pure) real number: scaling
* A (pure) imaginary number: rotation
* A complex number: scaling + rotation
